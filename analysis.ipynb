```python
# plot batch normalization graph similar to https://r2rt.com/implementing-batch-normalization-in-tensorflow.html

import matplotlib.pyplot as plt
%matplotlib inline

with open('logs/bn.csv', 'r') as f:
    acc_bn = []
    for l in f.readlines():
        a = float(l.split(',')[1])
        acc_bn.append(a)
        
with open('logs/no_bn.csv', 'r') as f:
    acc_no_bn = []
    for l in f.readlines():
        a = float(l.split(',')[1])
        acc_no_bn.append(a)

fig, ax = plt.subplots()

ax.plot(range(0,len(acc_no_bn),1),acc_no_bn, label='Without BN')
ax.plot(range(0,len(acc_bn),1),acc_bn, label='With BN')
ax.set_xlabel('Training steps')
ax.set_ylabel('Accuracy')
# ax.set_ylim([0.1,1])
ax.set_title('Batch Normalization Accuracy')
ax.legend(loc=4)
plt.show()
```
